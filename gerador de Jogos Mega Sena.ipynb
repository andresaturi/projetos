```python
from random import randint
from time import sleep
jogos = list()
numeros = list()

print('----GERADOR DE JOGOS MEGASENA----')
print('')
sleep(1)

q = int(input('Quantos jogos quer gerar?: '))
try:
  sorte = int(input('Número da sorte (0 caso não tenha): '))
  reves = int(input('Número do revés (0 Caso não tenha): '))
except:
  pass

for j in range(0, q):
    cont = 0
    if sorte > 0:
      numeros.append(sorte)
    
    while True:
        n = (randint(1, 60))
        if n not in numeros and n != reves:
            numeros.append(n)
            cont += 1
        if cont >= 6:
            break
    jogos.append(sorted(numeros[:]))
    numeros.clear()
print('')
print(f'Processando {q} jogos...')
sleep(3)
for c, v in enumerate(jogos):
    print(f'Jogo {c+1}: {jogos[c]}')
    sleep(1)
print('Finalizado.')

```
