```python
   # recepção dos 11 Digitos
temp = ''
while True:
    while True:
        cpf = str(input('Digite os Onze digitos do CPF: '))
        for n in cpf:
          if n.isnumeric() :
            temp += n
          else:
            pass
        cpf = temp
        temp = ''
        
        if len(cpf) == 11 and cpf != ('1'*11) and cpf != ('2'*11) and cpf != ('3'*11) and cpf != \
                ('4'*11) and cpf != ('5'*11) and cpf != ('6'*11) and cpf != ('7'*11)\
                and cpf != ('8'*11) and cpf != ('9'*11) and cpf != ('0'*11):
            break
        else:
            print('O CPF deve ter 11 Digitos não sequenciais...')
            print()

    # Listas
    cpflist = list()

    # Desempacotamento
    for c, d in enumerate(cpf):
        cpflist.append(d)
    cpflist.pop()
    cpflist.pop()
    # Digito 1
    soma = 0
    for c, d in enumerate(range(10, 1, -1)):
        dig = int(cpflist[c])
        soma += (dig * d)
    digito1 = 11-(soma % 11)
    if digito1 > 9:
        digito1 = 0
    # Digito 2
    segundo = cpflist.append(digito1)
    soma = 0

    for c, d in enumerate(range(11, 1, -1)):
        dig = int(cpflist[c])
        soma += (dig * d)
    digito2 = 11-(soma % 11)
    if digito2 > 9:
        digito2 = 0
    cpflist.append(digito2)
    novo = ''

    for c, d in enumerate(cpflist):
        novo += str(d)

    print()
    if cpf == novo:
        print('O CPF informado é VÁLIDO')        
    else:
        print('O CPF informado é INVÀLIDO,')
    print()

    cont = str(input('Continuar? [S/N]: ')).upper().strip()[0]

    if cont == 'N':
        break

print()
print('FIM DO PROGRAMA')
```
