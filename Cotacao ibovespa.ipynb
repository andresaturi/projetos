```python
import requests
from bs4 import BeautifulSoup
from time import sleep

url = 'https://br.investing.com/indices/bovespa'

media = list()
periodos = 21
tempo_grafico = 5

while True:
  response = requests.get(url)
  html = BeautifulSoup(response.content, 'html.parser')
  ibovespa = html.find('span', class_= 'text-2xl')

  ibov_float = float(ibovespa.text)
  media.append(ibov_float)

  if len(media) >= periodos:
    del media[0]

  resultado = (sum(media)/len(media))  

  if resultado > ibov_float:
    print('O Preço está acima da média')
  else:
    print('O preço está abaixo da média')
  
  sleep((tempo_grafico*60)-1)
```
